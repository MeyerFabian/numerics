```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
F = (np.random.rand(3,3)-.5)*.5+np.identity(3)

print(F)
```

```python
from mpl_toolkits.mplot3d import Axes3D;
```

```python
sampling_rate = 1/25;
dimension = int(1/sampling_rate)+1
x= np.empty([3,dimension*dimension*dimension])
```

```python
for i in range(0,dimension):
    for j in range(0,dimension):
        for k in range(0,dimension):
            x[:,i+j*dimension+k*dimension*dimension] = 2.0*np.array([sampling_rate*i,sampling_rate*j,sampling_rate*k])-1.0
    
```

```python
X=np.dot(F,x)
```

```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

ax.scatter(x[0,:],x[1,:],x[2,:])
ax2.scatter(X[0,:],X[1,:],X[2,:])


plt.show()
```
