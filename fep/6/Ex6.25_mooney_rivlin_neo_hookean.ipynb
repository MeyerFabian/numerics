```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def stress(C1,C2,p_stretch):
    y = p_stretch
    y2 = np.multiply(y,y)
    y3 = np.multiply(y2,y)
    return 2*(C1*(y-1/(y2))+C2*(1-1/(y3)))
```

```python
eng_strain = np.linspace(-0.5,3.0,num=100)
```

```python
princ_stretch = eng_strain + 1
```

```python
C1_neo = 100.0
C2_neo = 0.0
stress_neo = stress(C1_neo,C2_neo,princ_stretch)
```

```python
C1_mooney = 75.0
C2_mooney = 25.0
stress_mooney = stress(C1_mooney,C2_mooney,princ_stretch)
```

```python
plt.xlabel('eng. strain')
plt.ylabel('eng. stress')
plt.grid(True, which='both')
plt.axis([-.5, 3,-1000,1000])
plt.plot(eng_strain,stress_neo,label="Neo-Hookean")
plt.plot(eng_strain,stress_mooney,label="Mooney Rivlin")
plt.legend()
plt.show()
```
